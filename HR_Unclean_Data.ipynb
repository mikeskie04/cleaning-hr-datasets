```python
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/eyowhite/Messy-dataset/raw/main/messy_HR_data.csv")
df
```

```python
df["Name"] = df["Name"].str.title()
df
```

```python
df.isnull().sum()
```

####

I tried to use the mean of the column but i got an error because the data type is accepting a string values. So to check what are the inputted data's I used values_count to simply view the data that needed to be changed
In order for us to change the data type of the column to be an integer.

```python
df['Age'].value_counts()
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df
```

###

I changed the string value to none because i will convert the column into a int

```python
df['Age'] = df['Age'].fillna(30)
df
```

###

filling the NaN/null values with 30

```python
### Make the column int

df['Age'] = df['Age'].astype(int)
df
```

```python
### Change the value of SIXTY THOUSAND TO 60000
df['Salary'] = df['Salary'].replace({'SIXTY THOUSAND': 60000})
df
```

```python
### Using this again to make strings a nan/null values

df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df
```

```python
### changing the nan/null values in the salary to the mean

df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df.isnull().sum()
```

```python
### No Problem with gender column
df['Gender'].value_counts()
```

```python
df['Joining Date'].replace({'April 5, 2018': '2018-04-05'})
```

```python
df['Joining Date'] = pd.to_datetime(df['Joining Date'], errors='coerce').dt.date
df
```

```python
df['Joining Date'] = df['Joining Date'].fillna(' ')
df
```

```python
df['Email'] = df['Email'].fillna("no_email@example.com")
df.head(10)
```

```python
df['Salary'] = df['Salary'].astype(int)
df
```

```python
df['Phone Number'] = df['Phone Number'].fillna('000-000-0000')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gender = df['Gender'].value_counts()
sns.countplot(data=df, x='Gender', palette='pastel', hue='Gender')
plt.title('Gender Distribution in Company')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

```python
score_map = {'A': 4, 'B': 3, 'C': 2, 'D': 1}
df['Performance_Value'] = df['Performance Score'].map(score_map)
top10 = df.sort_values(by='Performance_Value', ascending=False).head(10)
top10 = df.sort_values(by=['Performance_Value', 'Salary'], ascending=[False, False]).head(10)
top10['Label'] = top10['Name'] + ' - ' + top10['Position']
plt.figure(figsize=(10,6))
sns.barplot(data=top10, x='Label', y='Performance_Value', palette='viridis', hue='Label')

plt.title('Top 10 Performing Employees (Based on Performance Grade)')
plt.xlabel('Employee (Name - Position)')
plt.ylabel('Performance Score (A=4, D=1)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
